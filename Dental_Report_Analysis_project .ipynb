```python
#While droping the rows with Null values , we use threshold value as 8 means keeping rows with 4 or less null values
#res1 = result.dropna(axis=0, how = 'any')
df1 = df.dropna(axis=0,  thresh = 8)
df1.shape
```

```python
df1.isnull().sum()
#df.shape
```

```python
def update_f1(f1,f2):
    if  f1==100:
        return f2
    else :
        return f1
```

```python
df1.f1 = df1.f1.fillna(100)
```

```python
arr1= [] 

for index, row in df1.iterrows():
    arr1.append(update_f1(df1['f1'][index],df1['f2'][index]))
arr1
```

```python
#replace column f1 with values of f2 only where it is null

df1.f1= arr1
```

```python
df1.isnull().sum()
```

```python
#drop rows with f1 as null values
df2 = df1.dropna(subset=['f1'], axis=0, how='any')
```

```python
df2.isnull().sum()
```

```python
df2.shape
```

```python
df3 = df2.drop(['sex','age','f8'],axis = 1)
df3.head(5)
```

```python
df4 = df3.interpolate(axis =1,method = 'linear')
df4.isnull().sum()
```

```python
df4['f8']=df2['f8']
df4['age']=df2['age']
df4['sex']=df2['sex']
#'sex','age','f8'
```

```python
df4.isnull().sum()
```

```python
# F8 have around 830 rows as null values 
# below are the mean Age value only for feature f8 
#H - 21     x > 20
#G - 19     17.5 < x < 20
#F - 16     15.5 < x < 17.5 
#E - 15     14.4 < x < 15.5
#D - 13.8   12.7 < x < 14.4
#C - 12     11 < x < 12.7
#B - 10     9.6 < x < 11 
#A - 9.2    9 < x < 9.6 
#1 - 8.7    X < 9 


def update_f8(f8,age):
    if  f8==100:
        if age < 9:
            return 1
        elif age < 9.6:
            return 2
        elif age < 11 :
            return 3
        elif age <12.7:
            return 4
        elif age < 14.4:
            return 5
        elif age<15.5:
            return 6
        elif age <17.5:
            return 7
        elif age <20 :
            return 8
        else :
            return 9
    else :
        return f8
```

```python
df4 = df4.fillna(100)
```

```python
arr2= [] 
for index, row in df4.iterrows():
    arr2.append(update_f8(df4['f8'][index],df4['age'][index]))
```

```python
df4['f8']= arr2
```

```python
df4.head(5)
```

```python
df4.shape
```

```python
# Preparing df for data min max normalisation of features f1 to f8
df5 = df4
df5 = df5.drop(['age'],axis = 1)
df5 = df5.drop(['sex'],axis = 1)
df5 = df5.drop(['id'],axis = 1)
```

```python

df5  = df5.astype(np.int64)
```

```python
# apply min max normalisation to df5 to convert all feature values between 0 and 1 
df5 = (df5 - df5.min()) / (df5.max() - df5.min())
df5.head()
```

```python
# PCA - dimentionality reduction 
from sklearn.decomposition import PCA
 
pca = PCA(n_components=4)
data_pca = pca.fit_transform(df5)
```

```python
# Drop f1 to f8 columns and concatenate PCA1 to PCA4 

df6=df5.drop(['f1','f2','f3','f4','f5','f6','f7','f8'],axis = 1)
df6['pca1']=data_pca[:,0]
df6['pca2']=data_pca[:,1]
df6['pca3']=data_pca[:,2]
df6['pca4']=data_pca[:,3]
```

```python
import numpy as np
df7 = df6
X = np.array(df7.drop(['age'], axis=1))
y = np.array(df7['age'])

```

```python
# Test Train data split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.svm import LinearSVR
from sklearn.tree import DecisionTreeRegressor 
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
models = [
    ('LinearRegression1', LinearRegression()),
    ('Lasso', Lasso()),
    ('Ridge', Ridge()),
    ('LinearSVM', LinearSVR()),
    ('Decision Tree', DecisionTreeRegressor()),
    ('RandomForestRegressor', RandomForestRegressor())    
   ## ('GaussianNB', GaussianNB()),
   ## ('SVM', SVC()),
    #####('KNN', KNeighborsClassifier()),
   ## ('DT', DecisionTreeClassifier()),
   ## ('Logi Regression',LogisticRegression()),
   
]
```

```python
## Random permutation cross-validator
DecisionTreeRegressorresult = []
RandomForestRegressorresult = []
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv = ShuffleSplit(n_splits=10,test_size=0.3, random_state=0)
#cross_val_score(LinearRegression(),X,y,cv=cv)
#result.append('DecisitionTreeRegressor')
DecisionTreeRegressorresult.append(cross_val_score(DecisionTreeRegressor(criterion='mse',splitter='best',min_samples_split=90 ),X,y,cv=cv))
print('DecisionTreeRegressorresult',np.mean(DecisionTreeRegressorresult))
#cross_val_score(DecisionTreeRegressor(criterion='mae'),X,y,cv=cv)
#cross_val_score(DecisionTreeRegressor(criterion='friedman_mse'),X,y,cv=cv)
#result.append('RandomForestRegressor')
RandomForestRegressorresult.append(cross_val_score(RandomForestRegressor(criterion='mse',n_estimators=10,min_samples_split=90,bootstrap='false'),X,y,cv=cv))
print('RandomForestRegressorresult',np.mean(RandomForestRegressorresult))
```

```python
#Best parameter check - observation : algo output changes at each execution 
# so both mse - mean squared error and friedman_mse are good for this dataset
DecisionTreeRegressorBestPara = []
RandomForestRegressorBestPara = []
cv= ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
parameters = {
                'criterion':['mse','friedman_mse','mae'],
                'splitter' : ['best','random'],
                'min_samples_split' : [70,75,80,85,90]
                #'min_samples_split' : [10,20,30,50,80,110,150,200]
                              
            }
#gs = GridSearchCV(DecisionTreeRegressor(criterion='mse',splitter='random',min_samples_split=20),parameters,cv = 10)
gs = GridSearchCV(DecisionTreeRegressor(),parameters,cv = 10)
gs.fit(X,y)
DecisionTreeRegressorBestPara.append(gs.best_params_)
print('DecisionTreeRegressorBestPara: ',DecisionTreeRegressorBestPara,' -- Best Score : ',gs.best_score_)

parameters = {
                'criterion':['mse','mae'],
                'bootstrap' : ['True','False'],
                'min_samples_split' : [40,50,60]
                #'min_samples_split' : [20,50,80,110,150,200]
                              
            }
#gs1 = GridSearchCV(RandomForestRegressor(criterion='mse',splitter='random',min_samples_split=20),parameters,cv = 10)
gs1 = GridSearchCV(RandomForestRegressor( n_estimators=10),parameters,cv = 10)
gs1.fit(X,y)
RandomForestRegressorBestPara.append(gs1.best_params_)
print('RandomForestRegressorBestPara: ',RandomForestRegressorBestPara,' -- Best Score : ',gs1.best_score_)

 
```
